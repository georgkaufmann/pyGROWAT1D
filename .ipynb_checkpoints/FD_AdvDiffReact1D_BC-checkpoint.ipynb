In this notebook, we add **boundary conditions** to the **transient advection-diffusion-reaction equation**.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pulse(x,mu=2.,sigma=1.,form='gauss'):
    """
    Initial pulse function centered at mu and having standard with sigma
    """
    if (form=='gauss'): 
        f = np.exp(-(x-mu)**2/(sigma)**2)
    if (form=='rect'):
        f = np.zeros_like(x)
        f[(x >= mu-sigma) & (x <= mu+sigma)] = 1
    return f
```

```python
xstep = 141
x,dx = np.linspace(-2,12,xstep,retstep=True)   # m
print(dx)
```

```python
def FD_AdvDiffReact1D(c,dx,dt,v,D,lamb,bc='flow'):
    """
    1D advection-diffusion-reaction operator
    FTCS forward time centered space
    """
    nx = c.shape[0]
    cold = np.copy(c)
    for i in range(nx):            
        # left boundary
        if (i==0):
            if (bc=='no' or bc=='noflow'):
                adv=0.;diff=0.;react=0.
            elif (bc=='flow'):
                adv   = -v*dt/2/dx * (-3*cold[i] + 4*cold[i+1] - cold[i+2]) 
                diff  = D*dt/(dx**2)*(2*cold[i] - 5*cold[i+1] + 4*cold[i+2] - cold[i+3])
                react = -lamb*cold[i]*dt
            elif (bc=='cyclic'):
                adv   = - v*dt/2/dx * (cold[i+1] - cold[nx-1]) 
                diff  = D*dt/(dx**2)*(cold[i+1] - 2*cold[i] + cold[nx-1])
                react = -lamb*cold[i]*dt
        # right boundary
        elif (i==nx-1):
            if (bc=='no' or bc=='noflow'):
                adv=0.;diff=0.;react=0.
            elif (bc=='flow'):
                adv   = -v*dt/2/dx * (cold[i-2] - 4*cold[i-1] + 3*cold[i]) 
                diff  = D*dt/(dx**2)*(-cold[i-3] + 4*cold[i-2] - 5*cold[i-1] + 2*cold[i])
                react = -lamb*cold[i]*dt
            elif (bc == 'cyclic'):
                adv   = - v*dt/2/dx * (cold[0] - cold[i-1]) 
                diff  = D*dt/(dx**2)*(cold[0] - 2*cold[i] + cold[i-1])
                react = -lamb*cold[i]*dt
        else:
            adv   = -v*dt/(2*dx) * (cold[i+1]-cold[i-1])
            diff  = D*dt/(dx**2)*(cold[i+1]-2*cold[i]+cold[i-1])
            react = -lamb*cold[i]*dt
        c[i] = cold[i] + adv + diff + react
    if (bc=='noflow'):
        c[0] = c[1]
        #c[-1] = c[-2]
        c[nx-1] = c[nx-2]
    return c
```

```python
def integralC(c,dx):
    nx = c.shape[0]
    intC = 0.
    for i in range(nx-1):
        intC += dx*(c[i]+c[i+1])/2
    return intC
```

```python
def plotC(x,cSaved,tSaved,aSaved):
    # plot initial condition and last time step
    fig,axs = plt.subplots(2,1,figsize=(10,7))
    axs[0].set_title('FTCS reaction ('+bc+' boundary condition)')
    axs[0].set_xlabel('x [m]')
    axs[0].set_ylabel('c')
    axs[0].grid(alpha=0.3)
    axs[0].plot(x,cSaved[:,0],label='t = 0 s',linewidth=1,linestyle='dashed',color=(0,0,0,1.0))
    for i in range(2,cSaved.shape[1],2):
        axs[0].plot(x,cSaved[:,i],lw=2,label='t = '+str(round(tSaved[i],1))+' s')
        axs[0].fill_between(x,cSaved[:,i],lw=3,edgecolor=(1,0,0,0.0), facecolor=(0,0,0,0.1))
    axs[0].legend(bbox_to_anchor=(1.20, 1),loc='upper right')

    axs[1].set_xlabel('t [s]')
    axs[1].set_ylabel('$\\int c(x) dx$')
    axs[1].set_ylim([1.3,2.0])
    axs[1].plot(tSaved,aSaved)

    plt.tight_layout()
    return
```

```python
def ADR1D(x,dx,timeStart=0.,timeEnd=16.,timeWrite=1.,v=0.5,D=0.1,thalf=1000.,bc='flow',form='gauss'):
    """
    """
    # get proper Courant criterion
    safety = 0.2
    dtadv = np.nan; dtdiff = np.nan
    if (D != 0):
        dtdiff = safety*dx**2/D
    if (v != 0):
        dtadv = safety*dx/v
    timeStep = np.nanmin([dtdiff,dtadv,0.1])
    print ('dx:                ',dx)
    print ('Courant time step: ',timeStep)
    # initial step
    lamb = 0.
    if (thalf <= 100): lamb = np.log(2)/thalf # 1/s
    time = timeStart
    c    = pulse(x,form=form)
    cSaved = np.copy(c)
    tSaved = np.array([0])
    aSaved = np.array([integralC(c,dx)])
    # time loop
    while (time < timeEnd):
        time += timeStep
        c = FD_AdvDiffReact1D(c,dx,timeStep,v,D,lamb,bc=bc)
        if (np.abs(time%timeWrite) < timeStep):
            cSaved = np.c_[cSaved,c]
            tSaved = np.append(tSaved,time)
            aSaved = np.append(aSaved,integralC(c,dx))
            #print(time,cSaved.shape)
    # plot
    plotC(x,cSaved,tSaved,aSaved)
    return tSaved,cSaved,aSaved
```

```python
bc='no'
tSaved,cSaved,aSaved = ADR1D(x,dx,v=0.5,thalf=1000,bc=bc,timeEnd=20)

bc='noflow'
tSaved,cSaved,aSaved = ADR1D(x,dx,v=0.5,thalf=1000,bc=bc,timeEnd=20)

bc='flow'
tSaved,cSaved,aSaved = ADR1D(x,dx,v=0.5,thalf=1000,bc=bc,timeEnd=20)

bc='cyclic'
tSaved,cSaved,aSaved = ADR1D(x,dx,v=0.5,thalf=1000,bc=bc,timeEnd=20)
```
